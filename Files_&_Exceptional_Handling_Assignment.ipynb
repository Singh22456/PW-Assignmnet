```python
#Q1. What is the difference between interpreted and compiled languages?
'''
A. A compiled language is one in which the entire program is translated into machine
code by a compiler before execution, producing an executable file that runs
directly on the system. This makes compiled languages generally faster and
allows errors to be detected during the compilation stage. In contrast,
an interpreted language is executed line-by-line by an interpreter at runtime,
without generating a separate executable file. This makes interpreted languages
easier to test and more flexible, but usually slower, with errors appearing only
when the specific line is run. Examples of compiled languages include C and C++,
while Python and JavaScript are common interpreted languages.
'''
```

```python
#Q2. What is exception handling in Python?
'''
A. Exception handling in Python is a mechanism that allows a program to respond
gracefully to errors that occur during execution. Instead of stopping the entire
program when an error happens, Python lets developers catch these errors using
the try, except, else, and finally blocks. The code inside the try block is
executed first, and if an error occurs, it is caught by the except block,
preventing the program from crashing. This makes programs more robust,
user-friendly, and easier to debug by handling issues such as invalid input,
file-not-found errors, or division by zero in a controlled manner.
'''
```

```python
#Q3. What is the purpose of the finally block in exception handling?
'''
A. The purpose of the finally block in exception handling is to run code no matter
what happens — whether an exception occurs or not. It is used for cleanup actions
like closing files, releasing resources, or ending connections.'''
```

```python
#Q4. What is logging in Python?
'''
Logging in Python is a way to record important events that happen while a program runs.
Instead of only printing messages to the screen, logging saves information like
errors, warnings, or status updates so developers can understand what the program
 did and diagnose problems later. Python’s built-in logging module helps track issues,
 debug programs, and keep detailed run-time records in the console or in log files.
'''
```

```python
#Q5. What is the significance of the __del__ method in Python?
'''
A. The __del__ method in Python is a destructor — it is automatically called when an object is about to be destroyed.
Its main purpose is to allow you to clean up resources (like closing files,
eleasing memory, or disconnecting from databases) before the object is removed from memory.
'''
```

```python
#Q7.  How can you handle multiple exceptions in Python?
'''
A. We can handel multiple exceptions in 3 ways

1. Use multiple except blocks
try:
    x = int("abc")
except ValueError:
    print("Value error occurred")
except TypeError:
    print("Type error occurred")


2. Handle multiple exceptions in a single except using a tuple
try:
    x = int("abc")
except (ValueError, TypeError):
    print("Some error occurred")

3. Use a generic except
try:
    x = int("abc")
except Exception as e:
    print("Error:", e)
```

```python
 #Q16. Demonstrate how to use memory profiling to check the memory usage of a small program.
 import tracemalloc

# Start memory tracking
tracemalloc.start()

# Small program for testing memory usage
def small_program():
    numbers = [i for i in range(10000)]
    return sum(numbers)

result = small_program()

# Get current and peak memory usage
current, peak = tracemalloc.get_traced_memory()

# Stop tracing
tracemalloc.stop()

print("Result:", result)
print(f"Current memory usage: {current} bytes")
print(f"Peak memory usage: {peak} bytes")
```

```python
#Q18. How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

# Create a logger
logger = logging.getLogger("my_logger")
logger.setLevel(logging.INFO)

# Create a rotating file handler
handler = RotatingFileHandler(
    "app.log",        # log file name
    maxBytes=1_000_000,  # rotate after 1MB
    backupCount=3        # keep 3 backup log files
)

# Create a logging format
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Add handler to logger
logger.addHandler(handler)

# Example log messages
for i in range(50000):
    logger.info(f"Log message number {i}")

print("Logging with rotation enabled. Check app.log and backup files.")
```

```python
#Q21. Write a Python program that reads a file and prints the number of occurrences of a specific word.
filename = "q1_output.txt"
word_to_find = "apple"

count = 0

with open(filename, "r") as file:
    text = file.read()
    words = text.split()      # split text into individual words
    count = words.count(word_to_find)

print(f"The word '{word_to_find}' occurs {count} times in the file.")
```

```python
#Q22. How can you check if a file is empty before attempting to read its contents?
import os

filename = "q1_output.txt"

# Check if file exists first
if os.path.exists(filename):

    # Check if file size is zero
    if os.path.getsize(filename) == 0:
        print("The file is empty.")
    else:
        with open(filename, "r") as file:
            content = file.read()
            print("File content:\n", content)

else:
    print(f"Error: The file '{filename}' does not exist.")
```

```python
#Q23.  Write a Python program that writes to a log file when an error occurs during file handling.
import logging

# Configure logging
logging.basicConfig(
    filename="file_error.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

filename = "q1_output.txt"

try:
    with open(filename, "r") as file:
        content = file.read()
        print(content)

except Exception as e:
    logging.error(f"Error while handling file '{filename}': {e}")
    print("An error occurred during file handling. Check file_error.log for details.")
```
